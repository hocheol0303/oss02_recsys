```python
import psycopg2

def query(text):
    conn = psycopg2.connect(
        host="localhost",
        database="testdb",
        user="testuser",
        password="",
    )
    cur = conn.cursor()
    cur.execute(text)
    conn.commit()
    cur.close()
    conn.close()

def show(text):
    conn = psycopg2.connect(
        host="localhost",
        database="testdb",
        user="testuser",
        password="",
    )
    cur = conn.cursor()
    cur.execute(text)
    rows = cur.fetchall()
    for row in rows:
        print(row)
    cur.close()
    conn.close()
```

```python
# user_data.json   
text='''
INSERT INTO user_data (id, gender, age, grade, likedChannel)
VALUES (0, 'female', 21, 'sophomore', ARRAY['A', 'E']);
'''

def make_user_query(user):
    if len(user['likedChannel']) > 0:
        text = f'''
        INSERT INTO users(id, gender, age, grade, likedChannel)
        VALUES ({user['userId']}, '{user['gender']}', {user['age']}, '{user['grade']}', ARRAY{user['likedChannel']});
        '''
        return text
    else:
        text = f'''
        INSERT INTO users(id, gender, age, grade)
        VALUES ({user['userId']}, '{user['gender']}', {user['age']}, '{user['grade']}');
        '''
        return text

with open('/Users/myserver/workspace/oss/model04_demographic_filtering/data04/user_data.json', 'r') as f:
    json_data = json.load(f)

for user in json_data:
    text = make_user_query(user)
    query(text)

json_data[0]
```

```python
# rating_train.csv   
text='''
INSERT INTO ratings (userId, itemId, rating)
VALUES (0, 2, 0.7);
'''

df = pd.read_csv('/Users/myserver/workspace/oss/model04_demographic_filtering/data04/rating_train.csv')

def make_rating_query(row):
    text = f'''
    INSERT INTO ratings (userId, itemId, rating)
    VALUES ({row['userId']}, {row['itemId']}, {row['rating']});
    '''
    return text

for index, row in df.iterrows():
    text = make_rating_query(row)
    query(text)
```

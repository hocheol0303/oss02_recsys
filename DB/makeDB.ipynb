```python
import psycopg2

def query(text):
    conn = psycopg2.connect(
        host="localhost",
        dbname="infotree",
        user="infotree", 
        password="info1234",
        port=5432
    )
    cur = conn.cursor()
    cur.execute(text)
    conn.commit()
    cur.close()
    conn.close()

def show(text):
    conn = psycopg2.connect(
        host="localhost",
        dbname="infotree",
        user="infotree", 
        password="info1234",
        port=5432
    )
    cur = conn.cursor()
    cur.execute(text)
    rows = cur.fetchall()
    lst = []

    for row in rows:
        lst.append(row)
    cur.close()
    conn.close()
    return lst
```

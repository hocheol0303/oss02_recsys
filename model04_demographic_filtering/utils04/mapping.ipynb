```python
import pandas as pd

# 유저의 인구통계적 정보를 매핑하는 데이터프레임을 생성하는 코드
df = pd.DataFrame(columns=['category', 'index'])

gender = ['gender_male', 'gender_female']
age = ['age_19-21', 'age_22-24', 'age_25-27', 'age_28-30', 'age_31-']
grade = ['grade_fresher', 'grade_sophomore', 'grade_junior', 'grade_senior', 'grade_eternal']
likedChannel = ['likedChannel_A', 'likedChannel_B', 'likedChannel_C', 'likedChannel_D', 'likedChannel_E']

categories = [gender, age, grade, likedChannel]

for category in categories:
    for i, item in enumerate(category):
        df.loc[len(df)] = [item, i]
df
```
